```python
# installing required libraries
import selenium 
from selenium import webdriver
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings("ignore")

from itertools import repeat
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.common import exceptions 
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.common.action_chains import ActionChains
import time
```

```python
# calling chrome driver
driver=webdriver.Chrome('chromedriver.exe')
```

```python
#defining function to scraping the required by keeping url as input
def attach(x,y):
    for w in x:
        y.append(w.text)
rating_list=[]
review_list=[]
short_review_list=[]
electronic_type=[]
def web_scrap(z,y):
    driver.get(z)
    pages=np.arange(1,300,1)
    for page in pages:
        if page==1:
            driver.find_element_by_xpath('//a[@class="_1LKTO3"]/span').click()
        else:
            try:
                driver.find_element_by_xpath('//a[@class="_1LKTO3"][2]/span').click()    
            except:    
                break
        time.sleep(5)
        try:
            rating=driver.find_elements_by_xpath('//div[@class="col _2wzgFH K0kLPL"]/div[1]/div[1]')
            attach(rating,rating_list)
            short_review=driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
            attach(short_review,short_review_list)
            review=driver.find_elements_by_xpath('//div[@class="col _2wzgFH K0kLPL"]/div[2]')
            attach(review,review_list)
            electronic_type.extend(repeat(y,10))
        except exceptions.StaleElementReferenceException:  
            pass
```

```python
df=pd.DataFrame(list(zip(rating_list,short_review_list,review_list,electronic_type)))
df.columns=['Rating','Summary','Full Review','gadget']
df.index+=1
```
